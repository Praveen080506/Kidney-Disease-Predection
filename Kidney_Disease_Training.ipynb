```python
import pandas as pd
dt = pd.read_csv("/content/kidney_disease_cleaned.csv")
dt = dt.dropna()
dt_encoded = pd.get_dummies(dt)
x = dt_encoded.drop("classification_notckd", axis=1)
y = dt_encoded["classification_notckd"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
input_dim = x_train.shape[1]  # this will be 36
model = Sequential()
model.add(Dense(64, input_shape=(input_dim,), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # adjust based on your target

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size=8, steps_per_epoch=x_train.shape[0] // 8)
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=8)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score  #Evaluation
accuracy_score(y_test,y_pred)
```
